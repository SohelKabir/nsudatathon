```python
def plot_activity(activity, data):

    fig, (ax0, ax1, ax2) = plt.subplots(nrows=3,
         figsize=(15, 10),
         sharex=True)
    plot_axis(ax0, data['timestamp'], data['x'], 'x')
    plot_axis(ax1, data['timestamp'], data['y'], 'y')
    plot_axis(ax2, data['timestamp'], data['z'], 'z')
    plt.subplots_adjust(hspace=0.2)
    fig.suptitle(activity)
    plt.subplots_adjust(top=0.90)
    plt.savefig(f'activity_intensity{activity}.png')
    plt.savefig(f'activity_intensity{activity}.pdf')
    plt.show()

def plot_axis(ax, x, y, title):

    ax.plot(x, y, 'r')
    ax.set_title(title)
    ax.xaxis.set_visible(False)
    ax.set_ylim([min(y) - np.std(y), max(y) + np.std(y)])
    ax.set_xlim([min(x), max(x)])
    ax.grid(True)

for activity in np.unique(df3['activity-code']):
    subset = df3[df3['activity-code'] == activity][:180]
    plot_activity(activity, subset)

```
